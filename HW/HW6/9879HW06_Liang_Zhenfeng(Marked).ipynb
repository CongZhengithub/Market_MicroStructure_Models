```python
%%R
# Get the price series
px <- tqBAC$PRICE
p <- as.numeric(px)
log_p <- log(p)
```

```python
%%R

# Define a function to calculate and make the 
plotEstimators <- function(log_p, m){
    
    # Compute the ZHOU estimator
    rv_ZHOU <- sapply(1:m,function(q){sqrt(ZHOU(log_p, q))})
    plot(1:m,rv_ZHOU[1:m],col="red",type="l",ylab="Square root of RV estimate",xlab="Subsampling frequency q")

    # Compute the TSRV estimator
    rv_TSRV <- sapply(1:m,function(q){sqrt(TSRV(log_p, q))})
    lines(1:m,rv_TSRV[1:m],main=NA, col="blue", type="l")

    # Compute the MSRV estimator
    rv_MSRV <- sapply(1:m,function(q){sqrt(MSRV(log_p, q))})
    lines(1:m,rv_MSRV[1:m],main=NA, col="purple", type="l")

    # Compute the KRVTH estimator
    rv_KRVTH <- sapply(1:m,function(q){sqrt(KRVTH(log_p, q))})
    lines(1:m,rv_KRVTH[1:m],main=NA, col="yellow", type="l")

    # Compute the KRVC estimator
    rv_KRVC <- sapply(1:m,function(q){sqrt(KRVC(log_p, q))})
    lines(1:m,rv_KRVC[1:m],main=NA, col="green", type="l")

    legend("topright", c("ZHOU","TSRV","MSRV","KRVTH","KRVC"),
           lty = c(1,1,1,1,1), col=c("red","blue","yellow","purple","green"))
}

```

```python
%%R

# The maximum subsampling frequency you would like to plot
m <- 20

# Part (a)
# Passing the log price series
plotEstimators(log_p,m)
```

## (a)

```python
%%R

# Part (a)
# Funtion to aggregate a tq object (such as tqBAC) to the nearest numberOfPerionds timescale. 
aggregateTQ <- function(tqObj, numberOfPeriods, timescale){
    agg = aggregatets(tqObj, k = numberOfPeriods, on = timescale)
    return (agg)
}
```

## (b)

```python
%%R

# Part (b)
# Function to compute the correlation between two high-frequency log return series (in tq format) 
# where the underlying prices have been aggregated to the nearest k seconds.

Cal_Corr <- function(logReturnSeries1, logReturnSeries2){
    return (cor(logReturnSeries1, logReturnSeries2))
}
```

## (c)

```python
%%R

# Part (c)
CalAggregatedCorr_k_seconds_frequency <- function(tqObj1, tqObj2, k){
    
    aggObj1 <- aggregateTQ(tqObj1, k, "secs")
    aggObj2 <- aggregateTQ(tqObj2, k, "secs")
    
    p1 <- as.numeric(aggObj1$PRICE)
    p2 <- as.numeric(aggObj2$PRICE)
    
    r1 <- diff(log(p1))
    r2 <- diff(log(p2))
    
    return (Cal_Corr(r1, r2))
} 

CorrWrapper <- function(k){
    return (CalAggregatedCorr_k_seconds_frequency(tqBAC, tqC, k))
}
```

```python
%%R

nn <- 600

# compute the correlation curve
CorrVec <- sapply(2:nn, CorrWrapper)

# Smooth the curve
SmoothedOne <- ksmooth(2:nn, CorrVec, bandwidth = 10)$y

plot(2:nn,CorrVec,col="red",type="l",ylab="Correlation",xlab="Sampling interval k", 
     main = "Correlation between the price from tqBAC and tqC")
lines(2:nn,SmoothedOne,main=NA, col="blue", type="l")

legend("topright", c("Original","Smoothed"), lty = c(1,1), col=c("red","blue"))

```

## (d)

My result is consistent with Epps effect. From the above picture, we could see when the sampling interval k decrease, which means sampling frequency increases, the correlation tends to zero.

```python
%%R

IntradayCorrelation <- CorrWrapper(2)
cat("The intraday correlation between BAC and C returns on 04-May-2012 is", IntradayCorrelation)
```

## The Hayashi-Yoshida estimator

### 3. (2 points) <font color='blue'> Score: 2/2 </font>
Show that when two time series $p^{(1)}$ and $p^{(2)}$ are synchronous, the Hayashi-Yoshida estimator coincides with the realized covariance estimator.

## 3

The Hayashi-Yoshida estimator

$$HY_t=\sum_{i \leq t}\,\sum_{j \leq t}\,(p^{(1)}_i-p^{(1)}_{i-1})\,(p^{(2)}_j-p^{(2)}_{j-1})\,v_{ij}$$
where

$$v_{ij}={\bf 1}_{[t_{i-1},t_i] \cap [t_{j-1},t_j] \neq \emptyset}$$

When they are synchronous, $v_{ij} = 1$, $HY_t=\sum_{i \leq t}\,\sum_{j \leq t}\,(\Delta p^{(1)}_i)\,(\Delta p^{(2)}_j)$, we divide it by $T$, and would have the relized covariance.

```python
%%R 

# Forecasting daily Realized volatility for DJI 2008 using the basic harModel: HARRV

data(realized_library); #Get sample daily Realized Volatility data
DJI_RV <- realized_library$Dow.Jones.Industrials.Realized.Variance; #Select DJI
DJI_RV <- DJI_RV[!is.na(DJI_RV)] #Remove NA's
DJI_RV <- DJI_RV['2008']

x <- harModel(data=DJI_RV , periods = c(1,5,22), RVest = c("rCov"), type="HARRV",h=1,transform=NULL)
plot(x)
print(summary(x))
```
